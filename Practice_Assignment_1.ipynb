Q1 :: Given a number n, print the multiplication table from 1 to 10 for n in a single line, separated by spaces.

Example : Input = 2
          Output = 2 4 6 8 10 ..

```python
n  = int(input("Enter a number: "))
for i in range(1, 11, 1):
    print(n*i, end= " ")
```

Q3 :: Write a Python program that takes a positive integer $x$ as input. Your code should print a sequence of perfect squares ($1, 4, 9, 16, \dots$) on a single line, separated by spaces.
       The sequence must stop the moment the next square in the series would exceed the input value x

Example : Input = 10
          Output = 1 4 9

```python
x = int(input("Enter a number: "))
i = 1
while i**2 <= x:
    print(i**2, end= " ")
    i += 1
```

Q4 :: Given a list of integers , return a new list with items decreased by 2
Use Functional Programming concepts and higher order functions


Example : Input = [1,2,3]
          Output = [0,1,2]



```python
numbers = [1,2,3,4,5]
result = []

def decrease(numbers):
    return list(map(lambda x: x-2, numbers))

result = decrease(numbers)
print(result)
```
